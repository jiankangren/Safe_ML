```python
import pandas as pd
from matplotlib import cm
import numpy as np
from math import floor, ceil
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.metrics import accuracy_score
from Preprocess import *
from plot import *
markers=['.','+','x','_',',','|','1','2','3','4','8','>','<']
cmap1 = plt.cm.RdBu
cmap2 = plt.cm.Set2
cmap3 = plt.cm.tab20c
# cmap3 = plt.cm.hot
C =ListedColormap (np.random.rand (20,3)).colors
```

```python
file=['./data/SARCOS/','train.mat']
data=load(file[0]+'mlpbox')
def chunkIt(seq, num):
    avg = len(seq) / float(num)
    out = []
    last = 0.0

    while last < len(seq):
        out.append(seq[int(last):int(last + avg)])
        last += avg

    return out

data=chunkIt(data,10)
print len(data)
```

```python
figure = plt.figure(figsize=(20, 20))
ax = plt.subplot(2, 1, 1)
ax.set_title("Squared Loss For NNs ",size=20)

ax.set_xlabel('Sample Groups',size=20)
ax.boxplot(data, 0, '')

randomDists = [ [str(i*10-10)+'%',str(i*10)+'/%']  for i in   xrange(1,11)]
xtickNames = plt.setp(ax, xticklabels=randomDists)
plt.setp(xtickNames, rotation=45, fontsize=20)
ax.set_ylabel('Squared Loss',size=20)


ax = plt.subplot(2, 1, 1)
ax.set_title("Squared Loss  for Groups of Testing Samples ",size=20)

ax.set_xlabel('Sample Groups',size=20)
ax.boxplot(data, 0, '')

randomDists = [ [str(i*10-10)+'%',str(i*10)+'/%']  for i in   xrange(1,11)]
xtickNames = plt.setp(ax, xticklabels=randomDists)
plt.setp(xtickNames, rotation=45, fontsize=20)
ax.set_ylabel('Squared Loss',size=20)



plt.tight_layout()
plt.show()
```
