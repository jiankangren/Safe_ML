```python
def chunkIt(seq, num):
    avg = len(seq) / float(num)
    out = []
    last = 0.0

    while last < len(seq):
        out.append(seq[int(last):int(last + avg)])
        last += avg

    return out
```

# SARCOS

### SARCOS_NNs

```python
figure = plt.figure(figsize=(20, 10))

file=['./data/SARCOS/','train.mat']
data=load(file[0]+'mlpbox')
data=chunkIt(data,10)

ax = plt.subplot(1, 1, 1)
ax.set_title("Squared Loss for NN ",size=20)
ax.set_xlabel('Sample Groups',size=20)
ax.boxplot(data, 0, '')
randomDists = [ [str(i*10-10)+'%',str(i*10)+'/%']  for i in   xrange(1,11)]
xtickNames = plt.setp(ax, xticklabels=randomDists)
plt.setp(xtickNames, rotation=45, fontsize=20)
ax.set_ylabel('Squared Loss',size=20)


# plt.tight_layout()
# plt.show()
```
